```python
df.loc[df['Loan_Amount_Term'] == 180.0, 'Term'] = 0
df.loc[df['Loan_Amount_Term'] == 360.0, 'Term'] = 1
df
```

```python
df2.loc[df2['Loan_Amount_Term'] == 180.0, 'Term'] = 0
df2.loc[df2['Loan_Amount_Term'] == 360.0, 'Term'] = 1
```

```python
df['R'] = df['Property_Area'] == 'Rural'
df['S'] = df['Property_Area'] == 'Semiurban'
df['U'] = df['Property_Area'] == 'Urban'
df[['Property_Area', 'R', 'S', 'U']].head()
```

```python
df2['R'] = df2['Property_Area'] == 'Rural'
df2['S'] = df2['Property_Area'] == 'Semiurban'
df2['U'] = df2['Property_Area'] == 'Urban'
```

```python
df
```

```python
features = ['Term', 'R', 'S', 'U']
features
```

```python
labels = 'Loan_Status'
labels
```

```python
X_train = df[features]
print(X_train.shape)
X_train.head()
```

```python
X_test = df2[features]
print(X_test.shape)
X_test.head()
```

```python
Y_train = df[labels]
print(Y_train.shape)
Y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=5)
```

```python
decision_tree.fit(X_train, Y_train)
Y_predict = decision_tree.predict(X_test)
Y_predict
```
